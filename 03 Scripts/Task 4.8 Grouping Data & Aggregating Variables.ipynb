```python
# Q5. Determine whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean', 'min', 'max']})
```

#### Loyal customers are currently paying less as compared to the regular and new customers

```python
# Q6. Create a spending flag for each user based on the average price across all their orders

ords_prods_merge['mean_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price'] < 10, 'spending_flag'] = 'Low spender' 
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price'] >= 10, 'spending_flag'] = 'High spender' 
```

```python
ords_prods_merge.head(10)
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna =False)
```

#### Instacart as much more number of low spenders as compared to the high spenders

```python
# Q7. Create an order frequency flag that marks the regularity of a userâ€™s ordering behavior according to the median in the â€œdays_since_prior_orderâ€ column.
#calculating median 

ords_prods_merge['frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
# Using loc to calculate frequency numbers

ords_prods_merge.loc[ords_prods_merge['frequency'] > 20,'Customer Type'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['frequency'] <= 20) & (ords_prods_merge['frequency'] > 10),'Customer Type'] = 'Regular customer' 
```

```python
ords_prods_merge.loc[ords_prods_merge['frequency'] <= 10,'Customer Type'] = 'Frequent customer'
```

```python
ords_prods_merge.head(20)
```

```python
ords_prods_merge['Customer Type'].value_counts(dropna = False)
```

```python
ords_prods_merge = ords_prods_merge[ords_prods_merge['frequency'].isnull() == False]
```
